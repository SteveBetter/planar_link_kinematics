```python
import numpy as np
import matplotlib.pyplot as plt

length_scale = 1.3

L1 = 70.0 * length_scale
L2 = 80.0 * length_scale
L3 = 80.0

sita1 = [0.0, 90.0]
sita2 = [0.0, -180.0]
sita3 = [0.0, -180.0]

sita1_step = 1.0
sita2_step = 1.0
sita3_step = 1.0

sita1_range = np.linspace(sita1[0], sita1[1], num = int( np.floor( np.abs(sita1[1] - sita1[0]) / sita1_step) + 1), endpoint = True)
sita2_range = np.linspace(sita2[0], sita2[1], num = int( np.floor( np.abs(sita2[1] - sita2[0]) / sita2_step) + 1), endpoint = True)
sita3_range = np.linspace(sita3[0], sita3[1], num = int( np.floor( np.abs(sita3[1] - sita3[0]) / sita3_step) + 1), endpoint = True)

x_ = []
y_ = []
alpha_ = []

epsilon = 1e-3

for sita1_d in sita1_range:
    for sita2_d in sita2_range:
        for sita3_d in sita3_range:
            x, y, alpha = plink.three_link_kin(sita1_d, sita2_d, sita3_d, L1, L2, L3, unit = "deg")
            
            (sita1_tmp, sita2_tmp, sita3_tmp), (sita1_2_tmp, sita2_2_tmp, sita3_2_tmp) = plink.three_link_ik(x,y,alpha, L1, L2, L3, unit = "deg")
            
            if np.abs(sita1_2_tmp - sita1_d) >= epsilon or np.abs(sita2_2_tmp - sita2_d) >= epsilon or np.abs(sita3_2_tmp - sita3_d) >= epsilon:
                print("sita1, sita2, sita3: "  + str(sita1_d) + ", " + str(sita2_d) + ", "+ str(sita3_d))
            
            x_.append(x)
            y_.append(y)
            alpha_.append(alpha)

plt.plot(y_, x_, ".")
#x_ticks = np.linspace(0, L1 + L2, (L1 + L2) / 10) 
#plt.xticks(x_ticks)
#plt.xlim(0, L1 + L2)

plt.gca().invert_yaxis()
plt.grid()
plt.show()
```
